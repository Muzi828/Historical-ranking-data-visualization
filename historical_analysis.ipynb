```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"name":"华农兄弟","value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
```

```python
ls_name = ["敬汉卿","王刚","山药","中二所","观视频"]
for name in ls_name:
    df_ = pd.DataFrame({"name":name,"value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
    df = pd.concat([df,df_])
df
```

```python
rng = np.random.RandomState(1)
df = pd.DataFrame(rng.randint(10,100,size=(20,10))).add_prefix("y")
#生成20行，10列的数据，并添加标签
print(df)
df_top = df.cumsum(axis=0)#按列进行求和  太重要了
print(df_top)
df_right = df.cumsum(axis=1)
print(df_right)
```

```python
df = pd.DataFrame({"name":"华农兄弟","value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
```

```python
df["value"]= df["value"].cumsum(axis=0)
df
```

```python
ls_name = ["敬汉卿","王刚","山药","中二所","观视频"]
for name in ls_name:
    df_ = pd.DataFrame({"name":name,"value":np.random.randn(30)*1000,"date":pd.period_range("2019/10/1","2019/10/30")})
    df_["value"]= df_["value"].cumsum(axis=0)
    df = pd.concat([df,df_])
df
```
